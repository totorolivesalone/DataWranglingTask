```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import os
from google.colab import drive
```

```python
df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Aminu_Assignment/AminuInputFile.xlsx",sheet_name="Sheet1")
df.head()
```

```python
#Task 1
```

```python
df=df.iloc[:, :14]
df.columns=['order_id', 'date', 'account_code', 'product_name', 'quantity','appointment_date', 'category', 'unit_price', 'treat_disc', 'retail_disc', 'base_price', 'gst', 'final_price', 'check_rt']
df.head() #removing extra columns and rename them for better data handling
```

```python
df.dtypes
```

```python
df=df[df['order_id'].notna()] #removing all rows with missing order_id as order_id is primary key
df.head(-6)
```

```python
string_columns = ['account_code', 'product_name', 'category']
for col in string_columns:
    df[col] = df[col].astype(pd.StringDtype()) #converting objects to strings
df.dtypes
```

```python
print(df['retail_disc'].unique()[:20]) #checking why retail_disc is object type instead of float
```

```python
df['retail_disc'] = df['retail_disc'].astype(str).str.replace(',', '', regex=False).str.strip()
df['retail_disc'] = df['retail_disc'].replace(['-', '–', '—', ''], np.nan)  # Replace various dash types
df['retail_disc'] = pd.to_numeric(df['retail_disc'], errors='coerce')
print(df['retail_disc'].unique()[:20])
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date']) #converting dates data
df['appointment_date'] = pd.to_datetime(df['appointment_date'], errors='coerce')
df.dtypes
```

```python
missing_product = df['product_name'].isna()
missing_product_count = missing_product.sum()
missing_account =df['account_code'].isna()
missing_account_count=missing_account.sum()
missing_category=df['category'].isna()
missing_category_count=missing_category.sum()
missing_unitPrice=df['unit_price'].isna()
missing_unitPrice_count=missing_unitPrice.sum()
print(f"Rows with missing product: {missing_product_count}")
print(f"Rows with missing account: {missing_account_count}")
print(f"Rows with missing category: {missing_category_count}")
#checking if there are any rows with missing product , acc and category. None , so not dropping any rows
```

```python
prod = df.loc[df['category'].isna(), 'product_name'].iloc[0]
# Get a non‑null category for that product (first available)
cat = df[df['product_name'] == prod]['category'].dropna().iloc[0]
# Fill all missing rows for that product with that category
df.loc[df['category'].isna(), 'category'] = cat
```

```python
#evaluating numeric columns missing vals
missing_quantity_count=df['quantity'].isna().sum()
print("Rows with missing quantity",missing_quantity_count)
missing_unitPrice_count=df['unit_price'].isna().sum()
print("Rows with missing Unit Price",missing_unitPrice_count)
missing_treatDisc_count=df['treat_disc'].isna().sum()
print("Rows with missing Treat Discount",missing_treatDisc_count)
missing_retailDisc_count=df['retail_disc'].isna().sum()
print("Rows with missing Retail Discount",missing_retailDisc_count)
```

```python
df['quantity']=df['quantity'].fillna(value=0)
df['unit_price']=df['unit_price'].fillna(value=0)
df['treat_disc']=df['treat_disc'].fillna(value=0)
df['retail_disc']=df['retail_disc'].fillna(value=0)
missing_quantity_count=df['quantity'].isna().sum()
print("Rows with missing quantity",missing_quantity_count)
missing_unitPrice_count=df['unit_price'].isna().sum()
print("Rows with missing Unit Price",missing_unitPrice_count)
missing_treatDisc_count=df['treat_disc'].isna().sum()
print("Rows with missing Treat Discount",missing_treatDisc_count)
missing_retailDisc_count=df['retail_disc'].isna().sum()
print("Rows with missing Retail Discount",missing_retailDisc_count)
#handling na values for numeric fields except order_id
```

```python
#now lets standardise base_price , gst and final_price fields
#compute base price = unit_price*quantity*(1-treat_disc)*(1-retaildisc)
df['base_price'] = df['unit_price'] * df['quantity'] * (1 - df['treat_disc']) * (1 - df['retail_disc'])
df['base_price'] = df['base_price'].round(2)
#  Calculating GST as 18% of rounded base_price
df['gst'] = (df['base_price'] * 0.18).round(2)
# Calculate final_price = base_price + GST
df['final_price'] = (df['base_price'] + df['gst']).round(2)
```
